```python
import os
import glob
```

```python
paths = os.listdir('./fig/images/')
# print(paths)
paths_lists = []
for i in paths:
    paths_list = './fig/images/'+ i
    paths_lists.append(paths_list)
    
paths_lists

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(paths_lists[idx])
    
    if img is None:
        continue
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:
        break
    
    idx += 1
    if idx  >= len(paths_lists):
        idx =0
        
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig2/puppy.bmp', cv2.IMREAD_COLOR)

img1 = img[200:400, 300:500]
img2 = img[200:400, 300:500].copy()

#circle(img, center, radius, color[,thickness[,lineType[,shift]]]) -> img
cv2.circle(img1, (100,100),50, (0,0,255), 4, cv2.LINE_AA)

cv2.imshow('img' , img)
cv2.imshow('img1' , img1)
cv2.imshow('img2' , img2)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('./fig2/airplane.bmp', cv2.IMREAD_COLOR)
mask = cv2.imread('./fig2/mask_plane.bmp', cv2.IMREAD_GRAYSCALE)
dst = cv2.imread('./fig2/field.bmp')

if src is None or mask is None or dst is None:
    print('image read failed')
    sys.exit()
    
#cv2.copyTo(src,mask,dst =None) ->dst
cv2.copyTo(src,mask,dst)
    
cv2.imshow('src',src)
cv2.imshow('mask',mask)
cv2.imshow('dst',dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
